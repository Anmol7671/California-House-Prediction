```python
import pandas as pd
from sklearn.tree import DecisionTreeRegressor
```

```python
df=pd.read_csv("Gurgaon_housing.csv")
```

```python
df.isnull().sum()
```

```python
x=df.iloc[:, :-1]
y=df.iloc[:,-1]
```

```python
y
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=0)
x_train
```

```python
import numpy as np
housing_num=x_train.drop(["Location","Property_Type","Furnishing"],axis=1)
housing_num_test=x_test.drop(["Location","Property_Type","Furnishing"],axis=1)
housing_num
from sklearn.impute import SimpleImputer
SI=SimpleImputer(strategy="median")
housing_num = housing_num.select_dtypes(include=[np.number])
num_final=SI.fit_transform(housing_num)
num_final_test=SI.transform(housing_num_test)
```

```python
num_final
```

```python
housing_num_imputed=pd.DataFrame(num_final,columns=housing_num.columns,index=housing_num.index)
housing_num_imputed
housing_num_imputed_test=pd.DataFrame(num_final_test,columns=housing_num_test.columns,index=housing_num_test.index)
housing_num_imputed_test
```

```python
from sklearn.preprocessing import StandardScaler
SC=StandardScaler()
housing_num_imputed_scaled=SC.fit_transform(housing_num_imputed)
housing_num_imputed_scaled_test=SC.transform(housing_num_imputed_test)
```

```python
housing_num_imputed_scaled=pd.DataFrame(housing_num_imputed_scaled,columns=housing_num_imputed.columns,index=housing_num_imputed.index)
housing_num_imputed_scaled_test=pd.DataFrame(housing_num_imputed_scaled_test,columns=housing_num_imputed_test.columns,index=housing_num_imputed_test.index)
```

```python
housing_num_imputed_scaled
housing_num_imputed_scaled_test
```

```python
#lets deal with CAtegorical Values
housing_cat=x_train[["Location","Property_Type","Furnishing"]]
housing_cat
housing_cat_test=x_test[["Location","Property_Type","Furnishing"]]
```

```python
from sklearn.preprocessing import OneHotEncoder
OHE=OneHotEncoder()
encoded_house_cat=OHE.fit_transform(housing_cat)
encoded_house_cat_test=OHE.transform(housing_cat_test)
```

```python
encoded_house_cat.toarray()
```

```python
encoded_house_cat1=pd.DataFrame(encoded_house_cat.toarray(),columns=OHE.get_feature_names_out(housing_cat.columns),index=housing_cat.index)
encoded_house_cat1_test=pd.DataFrame(encoded_house_cat_test.toarray(),columns=OHE.get_feature_names_out(housing_cat_test.columns),index=housing_cat_test.index)
```

```python
encoded_house_cat1
```

```python
encoded_house_cat1_test
```

```python
x_train_final=np.hstack([housing_num_imputed_scaled,encoded_house_cat1])
x_train_final
x_test_final=np.hstack([housing_num_imputed_scaled_test,encoded_house_cat1_test])
x_test_final
```

```python
                                                #OR
```

```python
num_features = housing_num.columns
cat_features = OHE.get_feature_names_out(housing_cat.columns)

all_features = np.concatenate([num_features, cat_features])

x_train_final_df = pd.DataFrame(
    x_train_final,
    columns=all_features,
    index=x_train.index
)
#test set
num_features = housing_num.columns
cat_features = OHE.get_feature_names_out(housing_cat.columns)

all_features = np.concatenate([num_features, cat_features])

x_test_final_df = pd.DataFrame(
    x_test_final,
    columns=all_features,
    index=x_test.index
)
```

```python
x_train_final_df
```

```python
x_test_final_df
```

```python
DTR=DecisionTreeRegressor(random_state=0)
DTR.fit(x_train_final_df,y_train)
```

```python
y_pred=DTR.predict(x_test_final_df)
print(y_pred)
```

```python
print(y_test)
```

```python
from sklearn.metrics import mean_squared_error, r2_score

mse = mean_squared_error(y_test, y_pred)
rmse=np.sqrt(mse)
r2 = r2_score(y_test, y_pred)

print("RMSE:", rmse)
print("R2 Score:", r2)
```
